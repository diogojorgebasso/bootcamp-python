```
#adicionando métodos para modificar os atributos
class Veiculo:
    def __init__(self, marca, modelo, tipo):
        self.marca = marca
        self.modelo = modelo
        self.tipo = tipo
        self.tamanho_tanque_combustivel = 14
        self.nivel_tanque = 0
 
    def abastecer(self):
        self.nivel_tanque = self.tamanho_tanque_combustivel
        print('Tanque está cheio.')
 
    def dirigir(self):
        print('O {} está em movimento.'.format(self.modelo))

    def adicionando_combustivel(self, novo_nivel_tanque):
        if novo_nivel_tanque <= self.tamanho_tanque_combustivel:
            self.nivel_tanque = novo_nivel_tanque
        else:
            print('Transbordou')

    def set_nivel_combustivel(self, gasolina):
        if (self.nivel_tanque + gasolina <= self.tamanho_tanque_combustivel):
            self.nivel_tanque += gasolina
            print('Combustível adicionado.')
        else:
            print('Não cabe mais no tanque.')

    def get_nivel_combustivel(self):
        return self.nivel_tanque
```

```
objeto_fiat_uno=Veiculo('Fiat','Uno','Hatch')
```

```
#adicionando mais cobustível
objeto_fiat_uno.set_nivel_combustivel(3)
```

```
print('Nivel atual de combustível: ',objeto_fiat_uno.get_nivel_combustivel())
```

```
#adicionando mais cobustível
objeto_fiat_uno.set_nivel_combustivel(10)
```

```
#Herança em Python
class VeiculoEletrico(Veiculo):
    def __init__(self, marca, modelo, tipo, cor):
        super().__init__(marca, modelo, tipo)
        self.tamanho_bateria = 85
        self.nivel_carga = 0
        self.cor=cor

        #adicionando novos métodos à classe pai
    def carregar(self):
        self.nivel_carga = 100
        print('O veículo está completamente carregado.')
 
    def abastecer(self):
        print('Este veículo não tem tanque de combustível!')
```

```
#instâncias como atributos
class Bateria:
    def __init__(self, tamanho=85):
        self.tamanho = tamanho
        self.nivel_carga = 0
 
    def get_autonomia(self):
        if self.tamanho == 85:
            return 260
        elif self.tamanho == 100:
            return 315
```

```
#Herança em Python
class VeiculoEletrico(Veiculo):
    def __init__(self, marca, modelo, tipo):
        super().__init__(marca, modelo, tipo)
        self.bateria=Bateria()

        #adicionando novos métodos à classe pai
    def carregar(self):
        self.bateria.nivel_carga = 100
        print('O veículo está completamente carregado.')
 
    def abastecer(self):
        print('Este veículo não tem tanque de combustível!')
```

```
#instanciando os objetos
veiculo_eletrico = VeiculoEletrico('Tesla', 'CyberTruck', 'Caminhao')
veiculo_eletrico.carregar()
print('A autonomia do veículo é: ',veiculo_eletrico.bateria.get_autonomia())
veiculo_eletrico.dirigir()
```

```
#armazenando os objetos em uma lista
from veiculos import Veiculo, VeiculoEletrico
frota_combustivel_fossil = []
frota_eletricos = []

#criando a frota de veiculos
for _ in range(100):
    veiculo = Veiculo('Honda', 'Civic', 'Sedan')
    frota_combustivel_fossil.append(veiculo)

#criando a frota de veiculos eletricos
for _ in range(50):
    e_veiculos = VeiculoEletrico('Nissan', 'Leaf', 'Hatch')
    frota_eletricos.append(e_veiculos)

#abastecendo os veículos a combustão
for veiculo in frota_combustivel_fossil:
    veiculo.abastecer()

#carregando os veículos eletricos
for e_veiculo in frota_eletricos:
    e_veiculo.carregar()
    
print('Veículos a Combustão: {}'.format(len(frota_combustivel_fossil)))
print('Veículos Eletricos: {}'.format(len(frota_eletricos)))
```
